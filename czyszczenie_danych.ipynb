```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_table("TitanicMess.tsv")
df = df.set_index(df.PassengerId)
del df["PassengerId"]
```

```python
df
```

```python
# Sprawdzamy czy istnieje w zbiorze kolumna tylko z jednÄ… wartoÅ›Ä‡iÄ… oraz unikalnoÅ›Ä‡ danych w kolumnach.
print(df.nunique())
```

```python
counts = df.nunique()
to_del = [i for i,v in enumerate(counts) if v == 1]
print(to_del)
df.drop(df.columns[to_del], axis=1, inplace=True)
```

```python
# Kolumna "ship" zostaÅ‚a usuniÄ™ta.
print(df.nunique())
```

```python
# Sprawdzamy unikalne wartoÅ›ci dla kolumny "Sex"
df["Sex"].unique()
```

```python
# Kolumna "Sex" posiada bÅ‚Ä™dne dane dla pÅ‚ci male i female, takie dane naleÅ¼ poprawiÄ‡ i uspÃ³jniÄ‡
for index, row in df.iterrows():
    if pd.isna(row["Sex"]):
        continue
    elif row["Sex"] in ["malef", "mal"]:
        df.at[index, 'Sex'] = 'male'
    elif row["Sex"] in ["fem", "femmale"]:       
        df.at[index, 'Sex'] = 'female'
```

```python
# Po poprawnieniu danych pÅ‚ci unikalne wartoÅ›ci w tej kolumnie wyglÄ…dajÄ… nastepujÄ…co
df["Sex"].unique()
```

```python
df["Embarked"].unique()
```

```python
# Kolumna "Embarked" posiada bÅ‚Ä™dne dane, wykonujemy identyczne operacji jak przy klumnie pÅ‚eÄ‡
for index, row in df.iterrows():
    if pd.isna(row["Embarked"]):
        continue
    elif row["Embarked"] in "So":
        df.at[index, 'Embarked'] = 'S'
    elif row["Embarked"] in "Co":       
        df.at[index, 'Embarked'] = 'C'
    elif row["Embarked"] in "Qe":       
        df.at[index, 'Embarked'] = 'Q'
```

```python
# Dla kaÅ¼dej z kolumn sprawdzamy iloÅ›Ä‡ brakujÄ…cych wartoÅ›ci
missingData = pd.DataFrame(df.isna().sum())
missingData.columns = ["IloÅ›Ä‡ brakÃ³w"]
missingData
```

```python
# KolumnÄ™ "Cabin" usuwany ze zbioru poniewaÅ¼ posiada bardzo duÅ¼o brakujÄ…cych wartoÅ›ci
df.drop('Cabin', axis='columns', inplace=True)

# Usuwamy wiersze dla kÃ³trych Embarked przyjumuje NaN
df = df.dropna(axis='index',subset=['Embarked'])

# Usuwamy wiersze dla kÃ³trych Age przyjumuje NaN
df = df.dropna(axis='index',subset=['Age'])
```

```python
missingData = pd.DataFrame(df.isna().sum())
missingData.columns = ["IloÅ›Ä‡ brakÃ³w"]
missingData
```

```python
# Oczyszoczny zbiÃ³r danych zostaÅ‚ pisany do pliku titanic_clean.csv
df.to_csv("titanic_clean.csv")
```
