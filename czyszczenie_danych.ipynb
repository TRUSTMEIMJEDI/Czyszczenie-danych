```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_table("TitanicMess.tsv")
df = df.set_index(df.PassengerId)
del df["PassengerId"]
```

```python
df
```

```python
# Sprawdzamy czy istnieje w zbiorze kolumna tylko z jedną wartośćią oraz unikalność danych w kolumnach.
print(df.nunique())
```

```python
counts = df.nunique()
to_del = [i for i,v in enumerate(counts) if v == 1]
print(to_del)
df.drop(df.columns[to_del], axis=1, inplace=True)
```

```python
# Kolumna "ship" została usunięta.
print(df.nunique())
```

```python
# Sprawdzamy unikalne wartości dla kolumny "Sex"
df["Sex"].unique()
```

```python
# Kolumna "Sex" posiada błędne dane dla płci male i female, takie dane należ poprawić i uspójnić
for index, row in df.iterrows():
    if pd.isna(row["Sex"]):
        continue
    elif row["Sex"] in ["malef", "mal"]:
        df.at[index, 'Sex'] = 'male'
    elif row["Sex"] in ["fem", "femmale"]:       
        df.at[index, 'Sex'] = 'female'
```

```python
# Po poprawnieniu danych płci unikalne wartości w tej kolumnie wyglądają nastepująco
df["Sex"].unique()
```

```python
df["Embarked"].unique()
```

```python
# Kolumna "Embarked" posiada błędne dane, wykonujemy identyczne operacji jak przy klumnie płeć
for index, row in df.iterrows():
    if pd.isna(row["Embarked"]):
        continue
    elif row["Embarked"] in "So":
        df.at[index, 'Embarked'] = 'S'
    elif row["Embarked"] in "Co":       
        df.at[index, 'Embarked'] = 'C'
    elif row["Embarked"] in "Qe":       
        df.at[index, 'Embarked'] = 'Q'
```

```python
# Dla każdej z kolumn sprawdzamy ilość brakujących wartości
missingData = pd.DataFrame(df.isna().sum())
missingData.columns = ["Ilość braków"]
missingData
```

```python
# Kolumnę "Cabin" usuwany ze zbioru ponieważ posiada bardzo dużo brakujących wartości
df.drop('Cabin', axis='columns', inplace=True)

# Usuwamy wiersze dla kótrych Embarked przyjumuje NaN
df = df.dropna(axis='index',subset=['Embarked'])

# Usuwamy wiersze dla kótrych Age przyjumuje NaN
df = df.dropna(axis='index',subset=['Age'])
```

```python
missingData = pd.DataFrame(df.isna().sum())
missingData.columns = ["Ilość braków"]
missingData
```

```python
# Oczyszoczny zbiór danych został pisany do pliku titanic_clean.csv
df.to_csv("titanic_clean.csv")
```
